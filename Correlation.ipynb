```python
import numpy as np
import pandas as pd
from pandas_datareader import data, wb
import datetime
```

```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2010, 1, 11)
```

```python
# Yahoo data public and accessible 
f = data.DataReader("AAPL", 'yahoo', start, end)
f.tail()
# seems ok
```

```python
# Download all three
stocks = ['AAPL', 'GOOG', 'MSFT']
prices1 = [data.DataReader(s, 'yahoo', start, end) for s in stocks]
```

```python
prices1[2]
# Microsoft
```

```python
close = pd.concat([p['Close'] for p in prices1], axis=1, keys = stocks)
close
```
